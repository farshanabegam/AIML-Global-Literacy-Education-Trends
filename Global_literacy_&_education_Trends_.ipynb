```python
df_gdp = pd.read_csv("https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})
df_gdp.head()
```

```python
df_literacy = pd.merge(
    df_adult,
    df_youth,
    on=['country', 'year'],
    how='inner'
)

df_literacy.head()
```

```python
df_literacy.shape
```

```python
df_gdp_schooling = pd.merge(
    df_gdp,
    df_school,
    on=['country', 'year'],
    how='inner'
)
df_gdp_schooling.head()
```

```python
df_literacy = pd.merge(
    df_literacy,
    df_illiteracy,
    on=['country', 'year'],
    how='inner'
)

df_literacy.head()
```

```python
df_final = pd.merge(
    df_literacy,
    df_gdp_schooling,
    on=['country', 'year'],
    how='inner'
)

df_final.head()
```

```python
df_final.shape
```

```python
df_final.columns
```

```python
df_final['region'] = df_final['owid_region']
```

```python
df_final = df_final.drop(columns=[
    'owid_region',
    'owid_region_x',
    'owid_region_y'
], errors='ignore')
```

```python
df_final = df_final.drop(columns=[
    'population_historical',
    'code',
    'code_x',
    'code_y'
], errors='ignore')
```

```python
df_final.head()
df_final.columns
```

```python
df_final.shape
```

```python
df_final
```

```python
#feature engineering
#Literacy Gender Gap
df_final['literacy_gender_gap'] = df_final['youth_male'] - df_final['youth_female']
```

```python
#Youth Literacy Average
df_final['youth_literacy_avg'] = (
    df_final['youth_male'] + df_final['youth_female']
) / 2
```

```python
#GDP per Schooling Year
df_final['gdp_per_schooling'] = df_final['gdp_per_capita'] / df_final['avg_schooling']
```

```python
#Education Index
df_final['education_index'] = (
    df_final['literacy_rate'] * df_final['avg_schooling']
) / 100
```

```python
df_literacy.to_sql("literacy_rates", conn, if_exists="replace", index=False)
df_illiteracy.to_sql("illiteracy_population", conn, if_exists="replace", index=False)
df_gdp_schooling.to_sql("gdp_schooling", conn, if_exists="replace", index=False)
```

```python
pd.read_sql("SELECT * FROM literacy_rates LIMIT 5;", conn)
```

```python
#SQL Queries
 #literacy_rates
#1.Get top 5 countries with highest adult literacy in 2020.
query1 = """
SELECT country, adult_literacy
FROM literacy_rates
WHERE year = 2020
ORDER BY adult_literacy DESC
LIMIT 5;
"""

df1 = pd.read_sql(query1, conn)


```

```python
df1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.barplot(data=df1, x=df1.columns[1], y="country")

plt.title("Top 5 Countries with Highest Adult Literacy (2020)")
plt.xlabel("Literacy Rate (%)")
plt.ylabel("Country")

plt.show()
```

```python
#2.Find countries where female youth literacy < 80%.
query2 = """
SELECT country, year, youth_female
FROM literacy_rates
WHERE youth_female < 80;
"""

df2 = pd.read_sql(query2, conn)
```

```python
df2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,7))

sns.barplot(
    data= df2,
    x="youth_female",
    y="country"
)

plt.title("Countries with Female Youth Literacy Below 80%")
plt.xlabel("Female Youth Literacy (%)")
plt.ylabel("Country")

plt.show()
```

```python
#3.Average adult literacy per continent (owid region).
query3 = """
SELECT owid_region, AVG(adult_literacy) AS avg_literacy
FROM literacy_rates
GROUP BY owid_region;
"""

df3 = pd.read_sql(query3, conn)
```

```python
df3
```

```python


import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

ax = sns.barplot(data=df3, x="avg_literacy", y="owid_region")

for i, v in enumerate(df3["avg_literacy"]):
    ax.text(v + 0.5, i, round(v,2), va='center')

plt.title("Average Adult Literacy by Region")
plt.show()
```

```python
#Join Queries
#10. List top 10 countries in 2020 with highest GDP per capita but lowest average years of schooling(less than 6).
query10 = """
SELECT country, gdp_per_capita, avg_schooling
FROM gdp_schooling
WHERE year = 2020 AND avg_schooling < 6
ORDER BY gdp_per_capita DESC
LIMIT 10;
"""

df10 = pd.read_sql(query10, conn)
```

```python
#11. Show countries where the illiterate population is high despite having more than 10 average years of schooling.
query11 = """
SELECT i.country,
       i.illiteracy_rate,
       g.avg_schooling
FROM illiteracy_population i
JOIN gdp_schooling g
ON i.country = g.country AND i.year = g.year
WHERE g.avg_schooling > 10
AND i.illiteracy_rate > 10;
"""

df11 = pd.read_sql(query11, conn)
```

```python
df11
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df11, x="avg_schooling", y="illiteracy_rate")

for i in range(len(df11)):
    plt.text(
        df11["avg_schooling"].iloc[i],
        df11["illiteracy_rate"].iloc[i],
        df11["country"].iloc[i],
        fontsize=8
    )

plt.title("Education vs Illiteracy Paradox")
plt.show()
```

```python
#12. Compare literacy rates and GDP per capita growth for a selected country over the last 20 years. (country of your choice)
query12 = """
SELECT l.year,
       l.adult_literacy,
       g.gdp_per_capita
FROM literacy_rates l
JOIN gdp_schooling g
ON l.country = g.country
WHERE l.country = 'India'
ORDER BY l.year;
"""
df12 = pd.read_sql(query12, conn)
```

```python
df12
```

```python
fig, ax1 = plt.subplots(figsize=(10,5))

# Literacy (left axis)
ax1.plot(df12["year"], df12["adult_literacy"], 'b-o')
ax1.set_xlabel("Year")
ax1.set_ylabel("Literacy Rate (%)")

# GDP (right axis)
ax2 = ax1.twinx()
ax2.plot(df12["year"], df12["gdp_per_capita"], 'r-o')
ax2.set_ylabel("GDP per Capita")

plt.title("Literacy vs GDP Growth (India)")
plt.show()
```

```python
#13. Show the difference between youth literacy male and female rates for countries with GDP per capita above $30,000 in 2020.
query13 = """
SELECT l.country,
       l.youth_male,
       l.youth_female,
       (l.youth_male - l.youth_female) AS gender_gap,
       g.gdp_per_capita
FROM literacy_rates l
JOIN gdp_schooling g
ON l.country = g.country AND l.year = g.year
WHERE g.gdp_per_capita > 30000
AND l.year = 2020;
"""

df13 = pd.read_sql(query13, conn)
```

```python
df13
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# sort for better view
df13_sorted = df13.sort_values("gender_gap")

plt.figure(figsize=(10,6))

sns.barplot(
    data=df13_sorted,
    x="gender_gap",
    y="country"
)

plt.title("Gender Gap in Youth Literacy (High GDP Countries - 2020)")
plt.xlabel("Male - Female Literacy Gap (%)")
plt.ylabel("Country")

plt.show()
```
